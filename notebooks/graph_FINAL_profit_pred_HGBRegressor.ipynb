```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'../data/model_final/predictii_profit_HGBRegressor_blind_2017_final.csv')

fig, ax = plt.subplots(figsize=(16, 6))
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

ax.plot(df.index, df['profit'], color='steelblue', linewidth=0.8, alpha=0.85, label='Profit Real')
ax.plot(df.index, df['final_pred_profit'], color='green', linewidth=0.8, alpha=0.85, label='Profit Prezis (final_pred_profit)')

ax.set_title('Comparatie: Profit Real vs Profit Prezis (Blind 2017 dataset)', color='black', fontsize=14, fontweight='bold', pad=12)
ax.set_xlabel('Index (inregistrari)', color='black', fontsize=11)
ax.set_ylabel('Profit', color='black', fontsize=11)
ax.tick_params(colors='black')
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['top'].set_color('#cccccc')
ax.spines['right'].set_color('#cccccc')
ax.grid(True, color='#e0e0e0', linewidth=0.5)
ax.legend(fontsize=11, facecolor='white', edgecolor='black', labelcolor='black')

plt.tight_layout()
plt.savefig('FINAL_model_graphs/comparatie_profit.png', dpi=150, facecolor='white')
plt.show()
```
